```python
df = df[['id','original_title','overview','genres']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['tags'] = df['overview'] + df['genres']
```

```python
df.head()
```

```python
new_data = df.drop(['overview','genres'],axis='columns')
```

```python
new_data
```

```python
new_data.drop('overview_genre',axis='columns')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000,stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

WE FIND HERE THE CLOSE DISTANCE BETWEEN THE DATA POINTS WHICH IS 2000 DATA POINTS TO SEE THE PREFERRENCE OF MOVIE TASTE FOR THIS WE USE :_
COSINE SIMILARITY FUNCTION 


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar  =  cosine_similarity(vector)
```

```python
similar
```

```python
new_data.head()
```

```python
new_data[new_data['original_title']=='Insurgent'].index[0]
```

```python
new_data.info()

```

```python
similar[2]
```

```python
distance = sorted(list(enumerate(similar[2])),reverse=True,key=lambda vector: vector[1])
```

```python
for i in distance[0:5]:
    print(new_data.iloc[i[0]].original_title)
```

```python
def recommand(movies):
    index = new_data[new_data['original_title']==movies].index[0]
    distance = sorted(list(enumerate(similar[index])),reverse=True,key=lambda vector: vector[1])
    for i in distance[0:]:
        if(new_data.iloc[i[0]].original_title != movies):
            print(new_data.iloc[i[0]].original_title)



    
```

```python
print(distance)
```

```python
recommand("Furious 7")
```
